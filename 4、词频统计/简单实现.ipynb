```python
import jieba  
import os
import os

folder_path = '../2、下载pdf年报/年报txt版本/'  # 文件夹路径

def read_txt_files(folder_path):
    txt_files = []
    for root, dirs, files in os.walk(folder_path):
        for file in files:
            if file.endswith('.txt'):
                file_path = os.path.join(root, file)
                txt_files.append(file_path)
    return txt_files

txt_files = read_txt_files(folder_path)
txt_files
```

```python
merge_txt = '' 

for file in txt_files:
    # file_path = os.path.join(txt_files, file)  # 构建文件路径
    with open(file, 'r', encoding='utf-8') as f:
        content = f.read()
        merge_txt += content  # 将读取的内容添加到字符串变量中
        # 在这里进行处理，例如词频统计等

# 从文件夹中读取停用词
stopwords_path = './stopwords/hit_stopwords.txt'
stopwords = [line.strip() for line in open(stopwords_path, 'r', encoding='utf-8').readlines()]  # 从文件中读取所有行保存到list列表中

stopwords.extend(['本行', '公司'])  # 自定义停用词列表
# 增添所有数字
for i in range(100):
    stopwords.append(str(i))



filtered_words = list(filter(lambda x: x not in stopwords, jieba.cut(merge_txt)))


word_freq = {}  # 存储词频的字典
for word in filtered_words:  
    if len(word) > 1:  
        word_freq[word] = word_freq.get(word, 0) + 1  
  
word_freq_sorted = sorted(word_freq.items(), key=lambda x: x[1], reverse=True)  


keywords = [word  for word, freq in word_freq_sorted[:1000]]  # 取前5个关键词

print(keywords)
```

```python
for word , freq in word_freq_sorted[:10000]:
    # print(word, freq) 
    # 把word和freq写入csv文件
    with open('keywords.csv', 'a', encoding='utf-8') as f:
        f.write(word + ' ' + str(freq) + '\n')
```

```python
# 把csv文件转换成excel文件
import pandas as pd
df = pd.read_csv('keywords.csv', sep=' ', header=None)
df.to_excel('keywords.xlsx', index=False, header=False)
```
